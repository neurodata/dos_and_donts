```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read the simuation results
m10_wilcoxon = pd.read_csv("results/m10_wilcoxon")
m100_wilcoxon = pd.read_csv("results/m100_wilcoxon")
m10_mannwhitney = pd.read_csv("results/m10_mannwhitney")
m100_mannwhitney = pd.read_csv("results/m100_mannwhitney")
```

```python
df1 = pd.concat([m10_wilcoxon, m10_mannwhitney])
df1['m'] = 10
df2 = pd.concat([m100_wilcoxon, m100_mannwhitney])
df2['m'] = 100

df = pd.concat([df1, df2])
df.head()
```

```python
# Plot the power as a function of the connection strength
sns.set()
ax = sns.relplot(
    x=r"$p_{ij}$",
    y="power",
    hue="test",
    col="m",
    data=df,
    legend="full",
    kind="line",
)
plt.savefig("experiment1.png", dpi=300)
plt.show()
```
